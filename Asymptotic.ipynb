# Asymptotic Solutions to Delay Differential Equation

This notebook explores the asymptotic solution to the delay differential equation:

$$h'(u) = -(\ln 2)(h(u) + 1)h(u - 1)$$

For large $u$, solutions are well-approximated by the linear equation:
$$h'(u) = -(\ln 2)h(u - 1)$$

The characteristic equation $\omega e^\omega = -\ln 2$ has multiple solutions. The most dominant one leads to the asymptotic solution:

$$A e^{-0.571623609u} \sin(1.084461157u + B)$$

where $A$ and $B$ are constants determined by initial conditions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Parameters from the asymptotic analysis
alpha = 0.571623609  # decay rate
beta = 1.084461157   # oscillation frequency
```

## Figure 4: Original Asymptotic Solution

Reproducing Figure 4 with $h(0) = 1$

```python
def asymptotic_solution(u, A, B):
    """Asymptotic solution: A * e^(-alpha*u) * sin(beta*u + B)"""
    return A * np.exp(-alpha * u) * np.sin(beta * u + B)

# Find A and B such that h(0) = 1
# At u=0: A * sin(B) = 1
# We choose B such that we get an oscillatory behavior
# Using the values from the paper that match Figure 4

# From the paper, we need h(0) = 1
# A * sin(B) = 1
# Let's compute A and B from the boundary condition

def equations(vars):
    A, B = vars
    # Condition: h(0) = 1
    eq1 = A * np.sin(B) - 1
    # We also need another condition - let's use continuity
    # For Figure 4, we can use the derivative constraint or just fit to get nice oscillations
    # Let's use: A * cos(B) = some reasonable value (this affects the initial slope)
    eq2 = A * np.cos(B)  # Set to 0 for a specific phase, or use another constraint
    return [eq1, eq2]

# Solve for A and B
# Initial guess
A0, B0 = 1.5, 0.5
solution = fsolve(equations, [A0, B0])
A_initial, B_initial = solution

print(f"Initial fit: A = {A_initial:.10f}, B = {B_initial:.10f}")
print(f"Check h(0) = {asymptotic_solution(0, A_initial, B_initial):.10f}")

# For Figure 4, let's use values that produce h(0) = 1
# A better approach: use A and B from the paper's Figure 4
# Let's try: h(0) = 1, and choose B to get the right phase
# From inspection of Figure 4, it appears to start at 1 and initially decrease

# Set up a finer calculation
# h(u) = A*e^(-alpha*u)*sin(beta*u + B) = 1 at u=0
# So: A*sin(B) = 1

# From Figure 4, the function starts at 1, goes negative around u=0.5-1
# Let's use the constraint that h(0) = 1 and find reasonable A, B
# Trying: if B = 0, then A = 1/sin(0) = undefined
# If B = pi/2, then A*sin(pi/2) = A = 1

A_fig4 = 1.0
B_fig4 = np.pi / 2

print(f"\nFigure 4 parameters: A = {A_fig4}, B = {B_fig4:.10f}")
print(f"Check h(0) = {asymptotic_solution(0, A_fig4, B_fig4):.10f}")
```

```python
# Plot Figure 4
u_plot = np.linspace(0, 10, 1000)
h_plot = asymptotic_solution(u_plot, A_fig4, B_fig4)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(u_plot, h_plot, 'r-', linewidth=2, label=f'$A e^{{-{alpha:.3f}u}} \\sin({beta:.3f}u + \\pi/2)$')
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=12)
ax.set_ylabel('$h(u)$', fontsize=12)
ax.set_title('Asymptotic Solution: $h\'(u) = -(\\ln 2)(h(u) + 1)h(u - 1)$', fontsize=12)
ax.legend(fontsize=11)
ax.set_xlim(0, 10)
ax.set_ylim(-0.4, 1.1)

plt.tight_layout()
plt.show()

print(f"Maximum value: {np.max(h_plot):.6f}")
print(f"Minimum value: {np.min(h_plot):.6f}")
```

## Exploring Different A Values

The amplitude parameter $A$ controls the initial height of the oscillation.

```python
fig, ax = plt.subplots(figsize=(12, 6))

A_values = [0.5, 0.8, 1.0, 1.5, 2.0]
B_fixed = B_fig4

for A in A_values:
    u_plot = np.linspace(0, 10, 1000)
    h_plot = asymptotic_solution(u_plot, A, B_fixed)
    ax.plot(u_plot, h_plot, linewidth=2, label=f'$A = {A}$')

ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=12)
ax.set_ylabel('$h(u)$', fontsize=12)
ax.set_title('Effect of Amplitude Parameter $A$ (with $B = \\pi/2$)', fontsize=12)
ax.legend(fontsize=11, loc='best')
ax.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```

## Exploring Different B Values (Phase)

The phase parameter $B$ controls the initial angle of the oscillation.

```python
fig, ax = plt.subplots(figsize=(12, 6))

B_values = [0, np.pi/6, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
A_fixed = 1.0

for B in B_values:
    u_plot = np.linspace(0, 10, 1000)
    h_plot = asymptotic_solution(u_plot, A_fixed, B)
    ax.plot(u_plot, h_plot, linewidth=2, label=f'$B = {B/np.pi:.2f}\\pi$')

ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=12)
ax.set_ylabel('$h(u)$', fontsize=12)
ax.set_title('Effect of Phase Parameter $B$ (with $A = 1.0$)', fontsize=12)
ax.legend(fontsize=11, loc='best')
ax.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```

## Exploring Different Decay Rates

The decay rate $\alpha = 0.571623609$ controls how quickly the oscillations decay to zero.

```python
def asymptotic_solution_params(u, A, B, alpha, beta):
    """Asymptotic solution with custom alpha and beta"""
    return A * np.exp(-alpha * u) * np.sin(beta * u + B)

fig, ax = plt.subplots(figsize=(12, 6))

alpha_values = [0.3, 0.4, 0.571623609, 0.7, 0.9]
beta_fixed = beta
A_fixed = 1.0
B_fixed = B_fig4

for a in alpha_values:
    u_plot = np.linspace(0, 10, 1000)
    h_plot = asymptotic_solution_params(u_plot, A_fixed, B_fixed, a, beta_fixed)
    ax.plot(u_plot, h_plot, linewidth=2, label=f'$\\alpha = {a:.3f}$')

ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=12)
ax.set_ylabel('$h(u)$', fontsize=12)
ax.set_title(f'Effect of Decay Rate $\\alpha$ (with $A = {A_fixed}$, $B = \\pi/2$, $\\beta = {beta_fixed:.3f}$)', fontsize=12)
ax.legend(fontsize=11, loc='best')
ax.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```

## Exploring Different Oscillation Frequencies

The frequency parameter $\beta = 1.084461157$ controls how fast the oscillation occurs.

```python
fig, ax = plt.subplots(figsize=(12, 6))

beta_values = [0.5, 0.8, 1.084461157, 1.5, 2.0]
alpha_fixed = alpha
A_fixed = 1.0
B_fixed = B_fig4

for b in beta_values:
    u_plot = np.linspace(0, 10, 1000)
    h_plot = asymptotic_solution_params(u_plot, A_fixed, B_fixed, alpha_fixed, b)
    ax.plot(u_plot, h_plot, linewidth=2, label=f'$\\beta = {b:.3f}$')

ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=12)
ax.set_ylabel('$h(u)$', fontsize=12)
ax.set_title(f'Effect of Frequency $\\beta$ (with $A = {A_fixed}$, $B = \\pi/2$, $\\alpha = {alpha_fixed:.3f}$)', fontsize=12)
ax.legend(fontsize=11, loc='best')
ax.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```

## Comparison: Multiple Parameter Combinations

Let's create a comprehensive comparison with various parameter combinations.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
u_plot = np.linspace(0, 10, 1000)

# Case 1: Original Figure 4
ax = axes[0, 0]
h_plot = asymptotic_solution_params(u_plot, 1.0, np.pi/2, alpha, beta)
ax.plot(u_plot, h_plot, 'r-', linewidth=2)
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=10)
ax.set_ylabel('$h(u)$', fontsize=10)
ax.set_title('Figure 4 (Original)', fontsize=11, fontweight='bold')
ax.set_xlim(0, 10)

# Case 2: Faster decay
ax = axes[0, 1]
h_plot = asymptotic_solution_params(u_plot, 1.0, np.pi/2, 0.9, beta)
ax.plot(u_plot, h_plot, 'b-', linewidth=2)
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=10)
ax.set_ylabel('$h(u)$', fontsize=10)
ax.set_title('Faster Decay ($\\alpha = 0.9$)', fontsize=11, fontweight='bold')
ax.set_xlim(0, 10)

# Case 3: Higher amplitude
ax = axes[1, 0]
h_plot = asymptotic_solution_params(u_plot, 2.0, np.pi/2, alpha, beta)
ax.plot(u_plot, h_plot, 'g-', linewidth=2)
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=10)
ax.set_ylabel('$h(u)$', fontsize=10)
ax.set_title('Higher Amplitude ($A = 2.0$)', fontsize=11, fontweight='bold')
ax.set_xlim(0, 10)

# Case 4: Higher frequency
ax = axes[1, 1]
h_plot = asymptotic_solution_params(u_plot, 1.0, np.pi/2, alpha, 2.0)
ax.plot(u_plot, h_plot, 'm-', linewidth=2)
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$u$', fontsize=10)
ax.set_ylabel('$h(u)$', fontsize=10)
ax.set_title('Higher Frequency ($\\beta = 2.0$)', fontsize=11, fontweight='bold')
ax.set_xlim(0, 10)

fig.suptitle('Asymptotic Solutions with Different Parameters', fontsize=13, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```
